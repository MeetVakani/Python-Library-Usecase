```python
df = pd.read_excel("BC April 2021 Financial.xls", header=6, usecols='B:S', skipfooter=1)
df.head()
```

```python
df.tail()
```

```python
df_trans = df.T
df_trans.head()
```

```python
df_new_trans = df_trans.rename(columns=df_trans.iloc[0]).drop(df_trans.index[0])
df_new_trans.head()
```

```python
df1 = df_new_trans.assign(Index=range(len(df_new_trans))).set_index('Index')
df1.head()
```

```python
df1.drop(columns=df1.columns[:3], 
        axis=1, 
        inplace=True)
df1.head()
```
